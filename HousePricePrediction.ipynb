```python
selected_columns = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement']

plt.figure(figsize=(12, 6))

for i, column in enumerate(selected_columns):
    if pd.api.types.is_numeric_dtype(new_data[column]):
        plt.subplot(2, 4, i + 1)  # Arrange the subplots in a 2x4 grid
        new_data.boxplot(column=column)
        plt.title(f'Box Plot of {column}')
        plt.suptitle('')
        plt.xlabel('')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```
